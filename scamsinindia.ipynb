```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Modified_Inclusive_Indian_Online_Scam_Dataset (1).csv')
df['year'] = pd.to_datetime(df['transaction_time']).dt.year
df['month'] = pd.to_datetime(df['transaction_time']).dt.month
grouped_data = df.groupby('month').agg({'amount': 'sum'}).reset_index()
plt.figure(figsize=(12, 6))
plt.bar(grouped_data['month'], grouped_data['amount'])
plt.xlabel('Month')
plt.ylabel('Total Amount Scammed')
plt.title('Total Amount Scammed Every Month')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.gca().yaxis.set_major_formatter('{x:.0f}')
for i, bar in enumerate(plt.gca().patches):
    plt.gca().text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                   str(int(bar.get_height())), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Modified_Inclusive_Indian_Online_Scam_Dataset (1).csv')
grouped_data = df.groupby('card_type').size().reset_index(name='count')
grouped_data = grouped_data.sort_values(by='count', ascending=False)
top_5_card_types = grouped_data['card_type'].head(5)
df_top_5 = df[df['card_type'].isin(top_5_card_types)]
grouped_data_top_5 = df_top_5.groupby('card_type').size().reset_index(name='count')
plt.figure(figsize=(12, 6))
plt.pie(grouped_data_top_5['count'], labels=grouped_data_top_5['card_type'], autopct="%1.1f%%")
plt.title('Most Frequently Used Card Types for Scams')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Modified_Inclusive_Indian_Online_Scam_Dataset (1).csv')
location_counts = df.groupby('location').size().reset_index(name='Count')
location_counts = location_counts.sort_values('Count', ascending=False)
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(location_counts['location'], location_counts['Count'], color='red', linewidth=2)
ax.set_xlabel('Location', fontsize=14)
ax.set_ylabel('Number of Scams', fontsize=14)
ax.set_title('Frequency of Scams in a Locations', fontsize=16)
for i, row in location_counts.iterrows():
    ax.annotate(str(row['Count']), xy=(row['location'], row['Count']), ha='center', va='bottom')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
path = '/content/Modified_Inclusive_Indian_Online_Scam_Dataset (1).csv'
df = pd.read_csv(path)
category_counts = df.groupby('purchase_category').size().reset_index(name='Count')
plt.figure(figsize=(12, 6))
plt.pie(category_counts['Count'], labels=category_counts['purchase_category'], autopct='%1.1f%%', startangle=90)
plt.title('Purchase Category Distribution', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
path = '/content/Modified_Inclusive_Indian_Online_Scam_Dataset (1).csv'
df = pd.read_csv(path)
unique_ages = df['customer_age'].value_counts().reset_index(name='Count')
plt.figure(figsize=(12, 6))
plt.bar(unique_ages['customer_age'], unique_ages['Count'], color='blue')
plt.xlabel('Customer Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Number of times a Age group is scammed', fontsize=16)
for i, bar in enumerate(plt.gca().patches):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             str(bar.get_height()), ha='center', va='bottom')
plt.show()
```
